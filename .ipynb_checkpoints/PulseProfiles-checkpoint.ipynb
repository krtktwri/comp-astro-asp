```python
# Parameters
R = 3
dtheta = 0.01
dphi = 0.01
psiMax = alpha2psiCorrected(np.pi/2, R) #Psi corresponding to alpha = 90 degrees

# Initializing arrays/lists
thetaRange = np.arange(dtheta, 2*np.pi, dtheta)
phiRange = np.arange(dphi, np.pi, dphi)
phaseRange = np.arange(0, 2*np.pi, 0.1)
F = []
```

```python
# Psi - between radial vector and line of sight vector at different locations on the surface
#i - obsNspin
def psi(theta, phi, c, phase): 
    
    i = c[1]
    w = phase
        
    n_psi_hat = (np.sin(i)*np.sin(w), np.sin(i)*np.cos(w), np.cos(i))                               #Unit vector towards lOS 
    r_hat = (np.sin(phi)*np.cos(theta), np.sin(phi)*np.sin(theta), np.cos(phi)) #Unit vector to a position on surface 
    
    return np.arccos(np.dot(n_psi_hat, r_hat)) #Angle between n_psi_hat and r_hat
```

```python
# Normalized Intensity (Polar hotspots spanning 20 degrees of constant intensity)
def I(theta, phi, c):
    
    eta_p = c[0]
    theta_p = np.pi/2
    r = 0.1 #radius in degrees
    
    if (np.abs(phi-eta_p)<= r and np.abs(theta-theta_p)<= r):
        return 10
    else:
        return 1
```

```python
# Flux per unit area (Eq. 18 - mb2011)
def flux_per_unitArea(theta, phi, c, phase):

    a = psi2alphaCorrected(psi(theta, phi, c, phase), R)   
    
    if a>np.pi/2:
        return 0
    else:
        return I(theta, phi, c) * (R * dtheta*dphi) * np.cos(a)
```

```python
def pulseProfile(phase, c): 
    F = []
   #Summing over all area patches on the surface
    for i in range(0, len(thetaRange)):
        for j in range(0, len(phiRange)):
            F.append(flux_per_unitArea(thetaRange[i], phiRange[j], phase, c))
        
    return sum(F)            
```

```python
c1 = np.deg2rad([20, 30])                #(Angle between magnetic and spin axis, angle between spin axis and vector towards observer)
fullRotation1 = pulseProfile(phaseRange, c1)

plt.plot(phaseRange, fullRotation1)
```

```python
c1 = np.deg2rad([20, 30])                #(Angle between magnetic and spin axis, angle between spin axis and vector towards observer)
fullRotation1 = pulseProfile(phaseRange, c1)

c2 = np.deg2rad([30, 60])                #(Angle between magnetic and spin axis, angle between spin axis and vector towards observer)
fullRotation2 = pulseProfile(phaseRange, c2)

c3 = np.deg2rad([60, 80])                #(Angle between magnetic and spin axis, angle between spin axis and vector towards observer)
fullRotation3 = pulseProfile(phaseRange, c3)

c4 = np.deg2rad([0,90])               #(Angle between magnetic and spin axis, angle between spin axis and vector towards observer)
fullRotation4 = pulseProfile(phaseRange, c4)
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2,2)
fig.suptitle("Pulsar Profiles")

ax1.plot(phaseRange/(2*np.pi), fullRotation1)
plt.xlabel('Phase')
ax1.set_ylabel('Flux')
ax1.set_title('Type 1')

ax2.plot(phaseRange/(2*np.pi), fullRotation2)
ax2.set_title('Type 2')

ax3.plot(phaseRange/(2*np.pi), fullRotation3)
ax3.set_ylabel('Flux')
ax3.set_xlabel('Phase')
ax3.set_title('Type 3')

ax4.plot(phaseRange/(2*np.pi), fullRotation4)
ax4.set_xlabel('Phase')
ax4.set_title('Type 4')
```
