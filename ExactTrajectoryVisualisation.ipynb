```python
import numpy as np
from matplotlib import pyplot as plt, patches
from modules.exact import trajectory
from modules.myconstants import *
```

### Computing trajectories

```python
# naming scheme - coordinate_impactParameter
x_2, y_2 = trajectory(2)
x_lmin, y_lmin = trajectory(lmin)
x_3, y_3 = trajectory(3)
x_4, y_4 = trajectory(4)
x_5, y_5 = trajectory(5)
```

### Plotting

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.plot(x_2, y_2, color='k', zorder=1)
ax.plot(x_lmin, y_lmin, color='r', zorder=1)
ax.plot(x_3, y_3, color='k', zorder=1)
ax.plot(x_4, y_4, color='k', zorder=1)
ax.plot(x_5, y_5, color='k', zorder=1)


circle1 = patches.Circle((0, 0), radius=1, color='black', zorder=100)
ax.add_patch(circle1)
ax.patch.set_visible(False)

ax.set_xlabel(r"X/$r_G$")
ax.set_ylabel(r"Y/$r_G$")

plt.xlim([-10, 10])
plt.ylim([-10, 10])
plt.title("Photon Orbits in Schwarzschild Spacetime")
plt.grid(alpha=0.3)

plt.gca().set_aspect('equal')
plt.savefig('fig/PhotonOrbits.png')
```
