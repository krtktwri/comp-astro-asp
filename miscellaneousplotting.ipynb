```python
# standard libraries
import numpy as np 
import matplotlib.pyplot as plt

# for parallelization
from multiprocessing import Pool    # To parallelize the observations
from functools import partial       # To pool functions with multiple arguments

# my modules
from modules.surfaceintegration import *
from modules.polarizationtransport import *
from modules.starcatalog import *
```

```python
#Graph Parameters
plt.rcParams['figure.figsize'] = 10, 7
plt.rc('text', usetex=False)
plt.rcParams.update({'font.size': 15,
                     'legend.fontsize': 15})
plt.rcParams['font.family'] = 'serif'
plt.rcParams['axes.labelsize'] = 15
plt.rcParams['axes.titlesize'] = 15
plt.rcParams['lines.linewidth'] = 2.5
```

```python
# plotting a 2D heat map of linear_polarization for different values E and theta
rad_range = np.arange(0, np.pi, 0.01)
E_range = np.logspace(-1, 3, len(rad_range))

# creating a meshgrid for E and phase values
E, theta = np.meshgrid(E_range, rad_range)

# creating a 2D array for storing the values of linear polarization
linear_polarization_array = np.zeros((len(E), len(theta)))

# calculating the values of linear polarization for different values of E and theta
for i in range(len(E)):
    for j in range(len(theta)):
        linear_polarization_array[i][j] = linear_polarization(theta[i][j], 1, E[i][j])
        
# plotting the 2D heat map
plt.figure(figsize=(10, 7))
plt.pcolormesh(theta/np.pi, E, linear_polarization_array, cmap='jet')
plt.colorbar(label='Linear Polarization')
plt.xlabel(r'Angle between emitted photon and local magnetic field [rad]')
plt.ylabel(r'Energy of emitted photon [keV]')
plt.title('NM Polarization Dependance on Photon Energy \n and Angle with Magnetic Field ($10^{12}$ G)')
plt.yscale('log')
plt.axhline(1, color='white', linestyle='--')
plt.axhline(10, color='white', linestyle='--')
plt.savefig('fig/NMPL_B12.png', dpi=300)
```

```python
# plotting a 2D heat map of linear_polarization for different values E and theta
rad_range = np.arange(0, np.pi, 0.01)
E_range = np.logspace(-1, 3, len(rad_range))

# creating a meshgrid for E and phase values
E, theta = np.meshgrid(E_range, rad_range)

# creating a 2D array for storing the values of linear polarization
linear_polarization_array0 = np.zeros((len(E), len(theta)))
linear_polarization_array1 = np.zeros((len(E), len(theta)))
linear_polarization_array2 = np.zeros((len(E), len(theta)))
linear_polarization_array3 = np.zeros((len(E), len(theta)))

# calculating the values of linear polarization for different values of E and theta
for i in range(len(E)):
    for j in range(len(theta)):
        linear_polarization_array0[i][j] = linear_polarization(theta[i][j], 0.1, E[i][j])
        linear_polarization_array1[i][j] = linear_polarization(theta[i][j], 1, E[i][j])
        linear_polarization_array2[i][j] = linear_polarization(theta[i][j], 10, E[i][j])
        linear_polarization_array3[i][j] = linear_polarization(theta[i][j], 100, E[i][j])
```

```python
# plotting the 2D heat maps for all four values of B in two rows
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 15))

ax1.pcolormesh(theta/np.pi, E, linear_polarization_array0, cmap='jet')
ax1.set_ylabel(r'Energy of emitted photon [keV]')
ax1.set_title('Magnetic Field ($10^{11}$ G)')
ax1.set_yscale('log')
ax1.axhline(1, color='white', linestyle='--')
ax1.axhline(10, color='white', linestyle='--')

firstplot = ax2.pcolormesh(theta/np.pi, E, linear_polarization_array1, cmap='jet')
ax2.set_title('Magnetic Field ($10^{12}$ G)')
ax2.set_yscale('log')
ax2.axhline(1, color='white', linestyle='--')
ax2.axhline(10, color='white', linestyle='--')


ax3.pcolormesh(theta/np.pi, E, linear_polarization_array2, cmap='jet')
ax3.set_xlabel(r'Angle between $\hat{k}$ and $\vec{B}$ [rad]')
ax3.set_ylabel(r'Energy of emitted photon [keV]')
ax3.set_title('Magnetic Field ($10^{13}$ G)')
ax3.set_yscale('log')
ax3.axhline(1, color='white', linestyle='--')
ax3.axhline(10, color='white', linestyle='--')

lastplot = ax4.pcolormesh(theta/np.pi, E, linear_polarization_array3, cmap='jet')
ax4.set_xlabel(r'Angle between $\hat{k}$ and $\vec{B}$ [rad]')
ax4.set_title('Magnetic Field ($10^{14}$ G)')
ax4.set_yscale('log')
ax4.axhline(1, color='white', linestyle='--')
ax4.axhline(10, color='white', linestyle='--')

# creating a colobar common to all plots
fig.subplots_adjust(right=0.8)
cbar_ax = fig.add_axes([0.85, 0.15, 0.05, 0.7])
fig.colorbar(lastplot, cax=cbar_ax, label='Normal Mode Linear Polarization $p_L$')
plt.savefig('fig/NMPL_all.png', dpi=300)
```
