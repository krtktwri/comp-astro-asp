```python
import numpy as np
from modules.polarizationtransport import *
from modules.starcatalog import *
```

```python
#Graph Parameters
plt.rcParams['figure.figsize'] = 10, 7
plt.rc('text', usetex=False)
plt.rcParams.update({'font.size': 15,
                     'legend.fontsize': 15})
plt.rcParams['font.family'] = 'serif'
plt.rcParams['axes.labelsize'] = 15
plt.rcParams['axes.titlesize'] = 15
plt.rcParams['lines.linewidth'] = 2.5
```

```python
########## 1. Load the data ##########
phase = np.loadtxt('simData/phaseRange.csv', delimiter=',')
simObs1 = np.loadtxt('simData/2NS14.csv', delimiter=',')
simObs2 = np.loadtxt('simData/4NS14.csv', delimiter=',')
simObs3 = np.loadtxt('simData/6NS14.csv', delimiter=',')
simObs4 = np.loadtxt('simData/8NS14.csv', delimiter=',')
simObs5 = np.loadtxt('simData/10NS14.csv', delimiter=',')
simObs6 = np.loadtxt('simData/12NS14.csv', delimiter=',')
simObs7 = np.loadtxt('simData/14NS14.csv', delimiter=',')

########## 2. Calculate the degree of linear polarization ##########
pL1 = simObs1[0]/simObs1[2]*100
pL2 = simObs2[0]/simObs2[2]*100
pL3 = simObs3[0]/simObs3[2]*100
pL4 = simObs4[0]/simObs4[2]*100
pL5 = simObs5[0]/simObs5[2]*100
pL6 = simObs6[0]/simObs6[2]*100
pL7 = simObs7[0]/simObs7[2]*100

########## 3. Create mock data by adding noise ##########
### Standard Deviation of 0.5
mock1 = pL1 + np.random.normal(0, 0.5, len(pL1))
mock2 = pL2 + np.random.normal(0, 0.5, len(pL2))
mock3 = pL3 + np.random.normal(0, 0.5, len(pL3))
mock4 = pL4 + np.random.normal(0, 0.5, len(pL4))
mock5 = pL5 + np.random.normal(0, 0.5, len(pL5))
mock6 = pL6 + np.random.normal(0, 0.5, len(pL6))
mock7 = pL7 + np.random.normal(0, 0.5, len(pL7)) 

### Standard Deviation of 1
mock8 = pL1 + np.random.normal(0, 1, len(pL1))
mock9 = pL2 + np.random.normal(0, 1, len(pL2))
mock10 = pL3 + np.random.normal(0, 1, len(pL3))
mock11 = pL4 + np.random.normal(0, 1, len(pL4))
mock12 = pL5 + np.random.normal(0, 1, len(pL5))
mock13 = pL6 + np.random.normal(0, 1, len(pL6))
mock14 = pL7 + np.random.normal(0, 1, len(pL7)) 
```

```python
# plotting a heatmap of the mock data
obsData = np.array([pL7, pL6, pL5, pL4, pL3, pL2, pL1])
mockData1 = np.array([mock7, mock6, mock5, mock4, mock3, mock2, mock1])
mockData2 = np.array([mock14, mock13, mock12, mock11, mock10, mock9, mock8])

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(14, 5))
ax1.imshow(obsData, cmap='hot', interpolation='nearest', extent = [0, 2*np.pi, 2, 14], aspect='auto')
ax1.set_ylabel("Photon Energy [KeV]")
ax1.set_xlabel("Spin Phase [rad]")
ax1.set_title("Simulated Data")
ax2.imshow(mockData1, cmap='hot', interpolation='nearest', extent = [0, 2*np.pi, 2, 14], aspect='auto')
ax2.set_xlabel("Spin Phase [rad]")
ax2.set_title("Gaussian Noise [$\sigma$ = 0.5]")
ax3.imshow(mockData2, cmap='hot', interpolation='nearest', extent = [0, 2*np.pi, 2, 14], aspect='auto')
ax3.set_xlabel("Spin Phase [rad]")
ax3.set_title("Gaussian Noise [$\sigma$ = 1]")
plt.savefig('fig/mockData.png', dpi=300, bbox_inches='tight')
```
