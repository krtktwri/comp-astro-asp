```python
from modules.polarizationtransport import *
from modules.starcatalog import *
from modules.surfacemap import *
import numpy as np
import matplotlib.pyplot as plt

# for parallelization
from multiprocessing import Pool    # To parallelize the observations
from functools import partial       # To pool functions with multiple arguments

phaseRange = np.arange(0, 2*np.pi, 0.1)
```

```python
cap1 = capDetails(10, [1, 0], 0.1)   # 10 degree cap, only ordinary radiation, 0.01 degree background radiation
cap2 = capDetails(20, [1, 0], 0.1)   # 20 degree cap, only ordinary radiation, 0.01 degree background radiation
cap3 = capDetails(30, [1, 0], 0.1)   # 20 degree cap, only ordinary radiation, 0.01 degree background radiation

cap4 = capDetails(2, [1, 0], 0.1)    # 15 degree cap, only ordinary radiation, 0.01 degree background radiation
cap5 = capDetails(2, [0, 1], 0.1)    # 15 degree cap, only extraordinary radiation, 0.01 degree background radiation
cap6 = capDetails(2, [0.5, 0.5], 0.1)# 15 degree cap, mixed radiation modes, 0.01 degree background radiation
```

```python
k4 = np.transpose([stokes_PolarCap(m, NS14, 1, cap4) for m in phaseRange])
k5 = np.transpose([stokes_PolarCap(m, NS14, 1, cap5) for m in phaseRange])
k6 = np.transpose([stokes_PolarCap(m, NS14, 1, cap6) for m in phaseRange])
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

ax1.plot(phaseRange, k4[0], label='Ordinary Mode')
ax1.plot(phaseRange, k5[0], label='Extraordinary Mode')
ax1.plot(phaseRange, k6[0], label='Mixed')
ax1.set_xlabel("Spin Phase [Radians]")
ax1.set_ylabel("Degree of Linear Polarization")

ax2.plot(phaseRange, k4[1], label='Ordinary Mode')
ax2.plot(phaseRange, k5[1], label='Extraordinary Mode')
ax2.plot(phaseRange, k6[1], label='Mixed')
ax2.set_xlabel("Spin Phase [Radians]")
```
